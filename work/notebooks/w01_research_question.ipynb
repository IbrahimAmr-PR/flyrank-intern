## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

Research Question: Which high-visibility content pages are currently declining or at risk of decay and should be prioritized for a content refresh?
Unit of Analysis: A single pseudonymized content item (page).  
The Decision: Improving the allocation of limited human editorial and rewriting capacity across thousands of existing pages.  
The Action: An editor or content creator reviews the prioritized page and performs a targeted content update or optimization.  
Cost of a Wrong Call:False Positive (Type I Error): Recommending a stable page for a refresh, which wastes valuable editorial hours on an unnecessary update.  
False Negative (Type II Error): Missing a truly declining page, leading to a permanent loss of search impressions, organic traffic, and business revenue.
